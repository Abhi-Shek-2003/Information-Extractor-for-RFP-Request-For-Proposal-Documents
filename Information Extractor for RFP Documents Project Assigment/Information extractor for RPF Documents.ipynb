```python
import os
import json
from PyPDF2 import PdfReader
from bs4 import BeautifulSoup
from transformers import pipeline

# Initialize Hugging Face QA pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad", framework="pt")
print("Pipeline loaded successfully!")

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

# Function to extract text from HTML
def extract_text_from_html(html_path):
    with open(html_path, "r", encoding="utf-8") as file:
        soup = BeautifulSoup(file, "html.parser")
        return soup.get_text(separator="\n")

# Hugging Face-based extraction
def extract_with_huggingface(context, question):
    result = qa_pipeline(question=question, context=context)
    return result['answer']

# Define fields to extract using Hugging Face
def extract_fields_with_huggingface(text):
    fields = {
        "Bid Number": extract_with_huggingface(text, "What is the Bid Number?"),
        "Title": extract_with_huggingface(text, "What is the Title of the bid?"),
        "Due Date": extract_with_huggingface(text, "When is the Due Date?"),
        "Bid Submission Type": extract_with_huggingface(text, "What is the Bid Submission Type?"),
        "Term of Bid": extract_with_huggingface(text, "What is the Term of Bid?"),
        "Pre Bid Meeting": extract_with_huggingface(text, "Is there a Pre Bid Meeting? If yes, when?"),
        "Installation": extract_with_huggingface(text, "What is the Installation requirement for the bid?"),
        "Bid Bond Requirement": extract_with_huggingface(text, "Is there a Bid Bond Requirement? If yes, what is it?"),
        "Delivery Date": extract_with_huggingface(text, "What is the Delivery Date for the bid?"),
        "Payment Terms": extract_with_huggingface(text, "What are the Payment Terms for the bid?"),
        "Any Additional Documentation Required": extract_with_huggingface(text, "Is any Additional Documentation Required?"),
        "MFG for Registration": extract_with_huggingface(text, "What is the MFG for Registration?"),
        "Contract or Cooperative to use": extract_with_huggingface(text, "What Contract or Cooperative is to be used for the bid?"),
        "Model_no": extract_with_huggingface(text, "What is the Model number of the product?"),
        "Part_no": extract_with_huggingface(text, "What is the Part number of the product?"),
        "Product": extract_with_huggingface(text, "What product is being bid on?"),
        "contact_info": extract_with_huggingface(text, "What is the contact information for the bid?"),
        "company_name": extract_with_huggingface(text, "What is the company name for the bid?"),
        "Bid Summary": extract_with_huggingface(text, "What is the summary of the bid?"),
        "Product Specification": extract_with_huggingface(text, "What is the Product Specification for the bid?")
    }
    return fields

# Process documents in a folder
def process_documents_with_llm(folder_path):
    structured_data = []
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if file_name.endswith(".pdf"):
            text = extract_text_from_pdf(file_path)
        elif file_name.endswith(".html"):
            text = extract_text_from_html(file_path)
        else:
            continue

        # Use Hugging Face-based extraction
        fields = extract_fields_with_huggingface(text)
        structured_data.append(fields)

    return structured_data

# Save structured data to JSON
def save_to_json(data, output_path):
    with open(output_path, "w", encoding="utf-8") as json_file:
        json.dump(data, json_file, indent=4, ensure_ascii=False)

# Main function
if __name__ == "__main__":
    folder = "/content/drive/MyDrive/Emplay - assignment/Emplay - assignment/Bid1"  # Replace with folder path
    output_file = "/content/drive/My Drive/output1.json"

    extracted_data = process_documents_with_llm(folder)
    save_to_json(extracted_data, output_file)

    print(f"Structured data saved to {output_file}")

if __name__ == "__main__":
    folder = "/content/drive/MyDrive/Emplay - assignment/Emplay - assignment/Bid2"  # Replace with folder path
    output_file = "/content/drive/My Drive/output2.json"

    extracted_data = process_documents_with_llm(folder)
    save_to_json(extracted_data, output_file)

    print(f"Structured data saved to {output_file}")



```
